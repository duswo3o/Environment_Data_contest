### 반복문 실행하기

```python
# 라이브러리 불러오기
import requests
import pprint
import json
# pandas import
import pandas as pd
from pandas.io.json import json_normalize

# Open api 주소
base_url = 'http://apis.data.go.kr/6260000/BusanCommercialHistoryService/getCommercialHistoryList?serviceKey=5rAWjzIvpYIix7CeROaSVtec%2FuBaBersjsC7N85UoyR1U0%2BY%2BkARPMjHPJaF%2FoGy3U4qVxsTlOHqfZCGSCqsvQ%3D%3D&pageNo={}&numOfRows=5000&resultType=json'

df_list = []

for n in range(1,57):
    print('pageNO :', n)
    url = base_url.format(n)

    # url 불러오기
    response = requests.get(url)

    # 데이터 값 출력
    contents = response.text

    #문자열을 json으로 변경
    json_ob = json.loads(contents)

    # 필요한 내용만 꺼내기
    body = json_ob['getCommercialHistoryList']['body']['items']['item']

    # Dataframe으로 만들기
    dataframe = json_normalize(body)

    # 컬럼명 변경
    dataframe.columns = ['현황', '상점명', '대분류', '중분류', '업종명', '주소', '개업일', '폐업일', '좌표']
    
    df_list.append(dataframe)

    # dataframe
```

```python
df_list
```

```python
df = pd.concat(df_list)
df.to_csv('부산상권점포이력서비스.csv', encoding='cp949')
```

```python
business = df[df['현황']!='폐업'].reset_index().drop(['index', '개업일', '폐업일'], axis=1)
business
```
